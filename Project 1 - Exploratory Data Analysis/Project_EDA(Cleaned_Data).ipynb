### Z-TEST

#### A  two Samples has been taken n1 and n2 is 500 and 1000 and there sample means are 31000 and 300000 where taken from there same population. Is there  standard devation is same or not. prove the claim. Take the population standard devation is  212474.35

Null_Hypothesis : H0 : There standard devation will same for two samples.

Alternate Hypothesis : H1 : There standard devation will not be same for two samples.

```python
data = df_1.sample(500)
```

```python
data
```

```python
sample_salary1 = data['Salary'].mean()
sample_salary1
```

```python
data1 = df_1.sample(1000)
```

```python
data1
```

```python
sample_salary2 = data1['Salary'].mean()
sample_salary2
```

```python
sample_salary1 = 31000
sample_salary2 = 30000
sample_size_n1 = 500
sample_size_n2 = 1000
population_std= 212474.36

```

```python
sample_size_n = sample_size_n1 + sample_size_n2
```

```python
sample_size_n
```

```python
def z_score(sample_salary1,sample_salary2,sample_size_n,population_std):
    num = sample_salary1 - sample_salary2
    den = population_std*(1/sample_size_n)**0.5
    return num/den
```

```python
z_stat = z_score(sample_salary1,sample_salary2,population_std,sample_size_n)
z_stat
```

```python
alpha = 0.05
p_value = 2 * (1 - norm.cdf(abs(z_stat)))
p_value
if (p_value < alpha):
    print("We have sufficient evidence to Reject Null Hypothesis")
else:
    print("Fail to reject null hypothesis")
```
